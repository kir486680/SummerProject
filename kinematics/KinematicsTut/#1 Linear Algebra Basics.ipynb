Now we are getting into rotations. 

How would we accomplish such rotation? This might look intimidating, but hang tight! We will unwind this step by step!



Guess what? We can now simplify our equation into a convenient matrix!

I will create a convenient rotate function to perform rotations

```python
def rotate_2D(Vn,Vo, radians):
    """Use numpy to build a rotation matrix and take the dot product."""

    c, s = np.cos(radians), np.sin(radians) #finding cos and sin of the angle
    j = np.matrix([[c, -s], [s, c]]) # We are building a rotational matrix
    m = np.dot(j, [Vn, Vo]) # taking a dot product 

    return float(m.T[0]), float(m.T[1])
rotate_2D(2,3, 0.5) # the output is our Vx, Vy
```

The next step would be to combine our rotation and translation matrices.


Thats great! However, as engineers we want to simplify things as much as possible. This is where an idea of the homogeneous matrix comes in. They allow us to combine rotational and translational matrices into a single matrix. 



```python
def homogenious_mult(Vn,Vo, Px, Py, radians):
    c, s = np.cos(radians), np.sin(radians) #finding cos and sin of the angle
    H = np.matrix([[c, -s, Px], [s, c, Py],[0,0,1]]) # We are building a rotational matrix
    m = np.dot(H, [Vn, Vo,1]) # taking a dot product
    return float(m.T[0]), float(m.T[1])
homogenious_mult(3,3,1,2,0.1)
```
