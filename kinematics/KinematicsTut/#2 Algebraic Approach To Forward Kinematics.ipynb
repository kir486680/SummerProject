This might loook intimidating, but we will carefully disect this image. We will only focus on the firts joint(highlighted in a red circle). What does "X1Y1 frame is at orientation..." means? Well, this simply means that in order to find coordinates of Vx1y1 in the X0Y0 frame, we need to multiply Vx1y1 by the corresponding rotation matrix. What is the corresponding matrix? The coresponding rotation matrix is simply 2x2 matrix that takes $\theta$ into account(we derived this matrix in the linear algebra basics notebook).
We will use the function that we created in the previous tutorial

```python
def rotate_2D(Vn,Vo, radians):
    """Use numpy to build a rotation matrix and take the dot product."""

    c, s = np.cos(radians), np.sin(radians) #finding cos and sin of the angle
    j = np.matrix([[c, -s], [s, c]]) # We are building a rotational matrix
    m = np.dot(j, [Vn, Vo]) # taking a dot product
```
